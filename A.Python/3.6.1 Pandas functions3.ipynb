```python
import numpy as np
import pandas as pd
```

```python
data = {'Name':['Jai','Anuj','Jai','Princi','Gaurav','Anuj','Princi','Abhi','Anuj'],
       'Age':[27,24,22,32,33,29,27,32,32],
        'Address':['Nagpur','Pune','Kanpur','Allahabad','Jodhpur','Kanpur','Allahabad','Aligarh','Pune'],
       'Qualification':['Msc','MA','MCA','PHD','BE','B.com','Msc','MA','B.com'],
       'Score':[24,24,35,45,53,47,50,52,60]}

df = pd.DataFrame(data)
df
```

# Apply function

```python
def upper_case(value):
    value = value.upper()
    return value

df['Qualification'] = df['Qualification'].apply(upper_case)
df
```

```python
def updated_age(x):
    return x+2

df['Age'] = df['Age'].apply(updated_age)
df
```

```python
df['Age'] = df['Age'].apply(lambda x:x+2)
df
```

```python
df['Address'].apply(lambda x: x if x == 'Pune' else np.NaN)
```

# Transform

```python
df['Age'] = df['Age'].transform(lambda x:x+2)
df
```

# And condition()

```python
df[(df['star_rating']>9)&(df['duration']>150)]
```

```python
df[(df['star_rating']>8)&(df['duration']>150)&(df['content_rating']=='R')]
```

# OR condition()

```python
df[(df['star_rating']>9)|(df['duration']>200)]
```

```python
df[((df['star_rating']>9)&(df['content_rating']=='R'))|((df['genre']=='Adventure')&(df['duration']>200))]
```

# isin function()

```python
list1 = ['R', 'PG-13', 'NOT RATED', 'PG', 'UNRATED']

df[df['content_rating'].isin(list1)]
df.loc[np.where(df['content_rating'].isin(list1))]
```

# Not operator

```python
df.genre.unique()
```

```python
list1 = ['R', 'PG-13', 'NOT RATED', 'PG', 'UNRATED']
list2 = ['Crime', 'Action', 'Drama', 'Western', 'Adventure', 'Biography',
       'Comedy', 'Animation', 'Mystery', 'Horror']

df[~(df['content_rating'].isin(list1)|(df['genre'].isin(list2)))]
```

# statistical functions on groupby

```python
df.groupby('genre').sum()
```

```python
df.groupby('genre').mean()
```

```python
df.groupby('genre')['duration'].mean().reset_index()
```

```python
df.groupby(['genre','content_rating'])['title'].count()
```

```python
df.groupby(['genre','content_rating'])['duration'].sum().reset_index().head(20)
```

```python
df.groupby('content_rating')['duration'].max().reset_index()
```

# agg function in groupby`

```python
df.groupby('content_rating')['star_rating'].agg((np.mean,np.median,np.sum,np.max,np.min)).reset_index()
```

```python
df.groupby('content_rating')['genre'].agg(pd.Series.mode).reset_index()
```

```python
df.groupby('genre').agg({'star_rating':'mean','content_rating':'count'})
```

# aggregate function()

```python
df.groupby('genre').aggregate({'star_rating':'mean','content_rating':'count'})
```

```python
df.groupby('content_rating')['genre'].aggregate(pd.Series.mode).reset_index()
```

# sort in groupby


```python
df.groupby('genre',sort=True).aggregate({'star_rating':'mean','content_rating':'count'})
```

```python
df.groupby('genre',sort=False).aggregate({'star_rating':'mean','content_rating':'count'})
```

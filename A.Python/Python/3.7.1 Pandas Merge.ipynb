```python
import pandas as pd
import numpy as np
```

# append()

### row wise append

```python
df1 = pd.DataFrame({'ASCII Nums':np.arange(65,71),
                  'char':[chr(i) for i in range(65,71)]})
df1
```

```python
df2 = pd.DataFrame({'ASCII Nums':np.arange(69,75),
                  'char':[chr(i) for i in range(69,75)]})
df2
```

```python
df1.append(df2)
```

```python
df1.append(df2,ignore_index=True)
```

```python
df3 = pd.DataFrame({'ASCII Nums':np.arange(69,75),
                  'character':[chr(i) for i in range(69,75)]})
df3
```

```python
df1.append(df3,ignore_index=True)
```

```python
df4 = pd.DataFrame({'ASCII Numbers':np.arange(69,75),
                  'character':[chr(i) for i in range(69,75)]})
df4
```

```python
df1.append(df4,ignore_index=True)
```

### append column wise

```python
df1 = df1.T
df4 = df4.T
display(df1,df4)
df1.append(df4).T
```

### concat along rows

```python
pd.concat([df1,df2],axis=0)     # by default axis is 0 which means concat along rows
```

```python
pd.concat([df1,df4],ignore_index=True)
```

### coluwise concat

```python
pd.concat([df1,df4],axis=1)
```

```python
pd.concat([df1,df3],axis=1)
```

```python
df5 = pd.DataFrame({'ASCII Nums':np.arange(69,74),
                  'character':[chr(i) for i in range(69,74)]})
df5
```

```python
pd.concat([df1,df5],axis=1)
```

```python
pd.concat([df1,df5],axis=0)
```

```python
pd.concat([df1,df5],axis=1,join='outer')
```

```python
pd.concat([df1,df5],axis=1,join='inner')
```

```python
pd.concat([df1,df5],axis=0,join='outer')
```

```python
pd.concat([df1,df5],axis=0,join='inner')
```

```python
NOTE: ` 1. when using inner join, keeping axis = 1, only those rows which are common in both dataframes will be visible.
        2. when using inner join, keeping axis = 0, only those columns which are common in both dataframes will be visible.
```

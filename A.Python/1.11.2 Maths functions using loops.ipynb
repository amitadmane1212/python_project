# Wap to find factorial of a number

```python
a = int(input())
f = 1

for i in range(1,a+1):    # 1,2,3,4,5
    f = f*i
    
f
```

# Fibonacci Series

```python
n = int(input())
fibo = []
a,b=0,1

for i in range(n):
    fibo.append(a)
    c=a+b
    a=b
    b=c
    
fibo
```

```python
n = int(input())
fibo = [0,1]

for i in range(n-2):    # 0,1,2
    fibo.append(fibo[i]+fibo[i+1])
    
fibo
```
